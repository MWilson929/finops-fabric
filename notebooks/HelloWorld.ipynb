## Section 3: Simple Function Definition
Test function creation and basic logic

```python
# Define a greeting function
def create_greeting(name, environment="Development"):
    """
    Create a personalized greeting message
    
    Args:
        name (str): Name to greet
        environment (str): Environment context
    
    Returns:
        str: Formatted greeting message
    """
    return f"Hello, {name}! Welcome to Microsoft Fabric {environment} environment."

# Define a deployment status function
def get_deployment_status(project, version, success=True):
    """
    Generate deployment status message
    
    Args:
        project (str): Project name
        version (str): Version number
        success (bool): Deployment success status
    
    Returns:
        str: Deployment status message
    """
    status = "✅ SUCCESS" if success else "❌ FAILED"
    return f"Deployment Status: {status}\nProject: {project}\nVersion: {version}"

# Define a lakehouse connection test function
def test_fabric_connectivity():
    """
    Simulate fabric lakehouse connectivity test
    
    Returns:
        dict: Connection test results
    """
    import datetime
    
    # Simulate connection test
    connection_test = {
        "timestamp": datetime.datetime.now().isoformat(),
        "lakehouse_name": "FinOpsHub",
        "connection_status": "Connected",
        "workspace_id": "$workspace.$id",  # Dynamic reference for fabric-cicd
        "lakehouse_id": "$items.Lakehouse.FinOpsHub.$id"  # Dynamic reference
    }
    
    return connection_test

print("✅ Functions defined successfully!")
```

## Section 4: Test Function Execution
Execute the defined functions and display results

```python
# Test greeting function
print("=== Testing Greeting Function ===")
greeting1 = create_greeting("Data Engineer")
greeting2 = create_greeting("FinOps Team", "Production")
print(greeting1)
print(greeting2)

print("\n=== Testing Deployment Status Function ===")
deployment_success = get_deployment_status("FabricCICD", "1.0.0", True)
deployment_failure = get_deployment_status("TestProject", "0.9.0", False)
print(deployment_success)
print("\n" + deployment_failure)

print("\n=== Testing Fabric Connectivity Function ===")
connectivity_results = test_fabric_connectivity()
print("Fabric Connection Test Results:")
for key, value in connectivity_results.items():
    print(f"  {key}: {value}")

print("\n🎉 All function tests completed successfully!")
print("This notebook is ready for Microsoft Fabric CI/CD deployment!")
```
